```python
from typing import Any, Dict


def extract_opuser_details(raw_details: Any) -> Dict[str, str]:
    details = {
        f["field_name"]: f.get("label")
        for f in raw_details["field_list"]
        if f.get("label") is not None and f.get("label") != ""
    }
    details["opuserid"] = raw_details["userid"]
    return details
```

```python
extracted_opuser_details = list(map(extract_opuser_details, opusers_details))
extracted_opuser_details = pd.DataFrame(extracted_opuser_details)
extracted_opuser_details
```
